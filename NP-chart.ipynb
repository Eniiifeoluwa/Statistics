```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import *
import math
```

```python
data = pd.DataFrame({'Fruit': ['Banana', 'Mango', 'Apple', 'Orange', 'Pawpaw', 'Pineapple'], 'Number in Basket': [100, 100, 100, 100, 100, 100], 'Overriped': [2, 4, 3, 5, 1, 1]})
```

```python
binomial_mean = round(np.mean(data['Overriped']), 3)
probability_mean = binomial_mean/ 100
```

```python
def lower():
    return round(binomial_mean - 3 * sqrt(binomial_mean * (1-probability_mean)), 3)


def upper():
    return round(binomial_mean + 3 * sqrt(binomial_mean * (1-probability_mean)), 3)
```

```python
plt.figure(figsize= (12, 8))
plt.annotate(xy=(4.7, upper() - 0.3), text = 'UPPER LIMIT', color = 'red')
plt.axhline(upper(), color = 'red')
plt.annotate(xy=(4.59, binomial_mean -0.3), text = 'CONTROL LINE',color = 'blue')
plt.axhline(binomial_mean, color = 'blue')
plt.annotate(xy= (4.7, lower()- 0.3), text = 'LOWER LIMIT', color = 'black')
plt.axhline(lower(), color = 'black')
for i, text in enumerate(data['Overriped']):
    plt.annotate(f'{text:.2f}', (i-0.1, text + 0.3), color = 'red')

plt.grid(alpha = 0.6, color = 'black', linewidth = 0.6)
plt.title('BINOMIAL MEAN CHART (NP-CHART)', fontsize = 24, color = 'green', fontname = 'Times New Roman', fontweight = 'bold')

plt.yticks((-1.5, -0.5, 0.5, 1.5, 3.5, 4.5, 5.5, 6.5, lower() , binomial_mean, upper()))
sns.lineplot(x = data['Fruit'], y = data['Overriped'], color = 'green', marker = 'o');
```
